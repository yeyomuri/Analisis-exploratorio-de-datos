```python
import pandas as pd
import altair as alt
```

```python

Chicago_data = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/Traffic_Crashes1.csv'
token = 'pk.eyJ1IjoieWV5b211cmkiLCJhIjoiY2t4YjJtZTVjMnNmczJucG1qMjE0MWtqbiJ9.qALr-E9KAB0wS88KSi19uQ'

df = pd.read_csv(Chicago_data)
df['CRASH_DATE'] = df['CRASH_DATE'].apply(lambda x: pd.to_datetime(x,errors='coerce', utc=True))
```

```python
df.head(2)
```

```python
report1 = df.groupby(['LIGHTING_CONDITION', 'REPORT_TYPE', 'CRASH_HOUR']).agg({'NUM_UNITS': 'sum'})
report1 = report1.reset_index()
report1
```

```python
alt.Chart(report1).mark_bar().encode(
    x = 'LIGHTING_CONDITION',
    y = 'NUM_UNITS',
    color = 'REPORT_TYPE'
).properties(width = 220)
```
