```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
cancer_data = df.data
cancer_name = df.feature_names
```

```python
df_cancer = pd.DataFrame(cancer_data, columns = cancer_name)
```

```python
df_cancer.info()
```

```python
df_cancer_test = df_cancer.iloc[:, :6]
```

```python
corr_matrix = df_cancer_test.corr()
```

```python
sns.heatmap(corr_matrix)
```

```python
corr_values = corr_matrix.unstack()
```

```python
corr_values.sort_values(kind = 'quicksort')
```

```python
sns.pairplot(df_cancer_test)
```
